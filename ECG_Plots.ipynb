```python
import pandas as pd
```

```python
traindata=pd.read_csv("/content/gdrive/MyDrive/ECGData/mitbih_train.csv",header=None)
```

```python
import plotly.express as px
```

```python
i=0
isample=traindata[traindata[187]==i].sample(n=5, random_state=42)
plotdf=pd.DataFrame(columns=["Patient","Time","ECG Value"])
for rowi in range(5):
  for coli in range(187):
      plotdf.loc[plotdf.shape[0]]=[rowi,coli,isample.iloc[rowi][coli]]
fig = px.line(plotdf, x="Time", y="ECG Value",color="Patient", title="Normal")
fig.show()
```

```python
#Train Acc
newplotdf=pd.DataFrame(columns=["Model","Epochs","Train Accuracy"])
for model in ["FNN","CNN","LSTM","HYBRID"]:
  metricdf=pd.read_csv("/content/gdrive/MyDrive/ECGData/PlotData/"+model.lower()+"_plotdata.csv")
  for ind,row in metricdf.iterrows():
    newplotdf.loc[newplotdf.shape[0]] =[model,row["Epoch"],row["Train Acc"]]
fig = px.line(newplotdf, x="Epochs", y="Train Accuracy",color="Model", title="Train Accuracies by Epochs")
fig.show()
```

```python
#Val Acc
newplotdf=pd.DataFrame(columns=["Model","Epochs","Val Accuracy"])
for model in ["FNN","CNN","LSTM","HYBRID"]:
  metricdf=pd.read_csv("/content/gdrive/MyDrive/ECGData/PlotData/"+model.lower()+"_plotdata.csv")
  for ind,row in metricdf.iterrows():
    newplotdf.loc[newplotdf.shape[0]] =[model,row["Epoch"],row["Test Acc"]]
fig = px.line(newplotdf, x="Epochs", y="Val Accuracy",color="Model", title="Val Accuracies by Epochs")
fig.show()
```
